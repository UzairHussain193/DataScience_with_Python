## Data visualization on IRIS Dataset

```python
#  importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing dataset
phool = sns.load_dataset("iris")
phool
```

```python
phool.info()
```

```python
# checking null values
phool.isnull().sum()
```

```python
# plotting
phool.species.value_counts().plot(kind="bar")
```

```python
sns.countplot(x="species",data=phool)
```

```python
sns.countplot(x="sepal_length",hue="species",data=phool)
```

```python
phool.columns
```

```python
plt.scatter(data=phool,x="sepal_length",y="species")
```

```python
sns.scatterplot(data=phool,x="sepal_width",y="sepal_length",hue="species")
```

```python
sns.barplot(data=phool,x="petal_length",y="sepal_width",hue="species")
```

```python
sns.pairplot(phool)
```

```python
phool.hist()
```

```python
phool.hist(bins=10,figsize=(10,8),grid=False)
```

```python
sns.boxplot(data=phool,x="sepal_width",y="sepal_length",hue="species")
```

```python
sns.boxenplot(data=phool,x="sepal_width",y="sepal_length",hue="species")
```

```python
# grid plot usinf facetgrid
g = sns.FacetGrid(phool,col="species",hue="sepal_width")
g.map(plt.scatter,"sepal_length")
```

```python
# grid plot usinf facetgrid

g = sns.FacetGrid(phool,  hue="species",col='petal_length', margin_titles=True,
              palette={1:"black", 0:"red"})
# g = g.map(plt.scatter,"petal_length","petal_width",edgecolor="w").add_legend()

```
