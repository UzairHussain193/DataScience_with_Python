### what every column means here in titanic sns dataset?
The columns meanings from Titanic SNS Dataset include :

1. Survived - Whether passenger survived or did not (0 = No; 1 = Yes).
2. Pclass - The class of travel ordered by priority on board(1=Upper Class Passengers, 2=Middle Class Passengers ,3=Lower Class Passengers ).
3. Name - Full name of each person.
4. Sex - Gender of each individual. Male/Female values only .
5. Age - Age range within which people were grouped together based upon their age at time of traval.
6. SibSp - Total number of siblings/spouses who were accompanying with them while travelling.
7. Parch -Total number of parents \ children who were accompanying with them while travelling.
8. Ticket - Unique ticket number assigned to each person during the trip.
9.  Fare - Amount paid by each person for his\her journey.
10. Cabin - Location where cabins were allocated according to availability. If no available then marked 'N'.
11. Embarked - Port of embarkation either C = Cherbourg ; Q = Queenstowe ; S = Southampton
